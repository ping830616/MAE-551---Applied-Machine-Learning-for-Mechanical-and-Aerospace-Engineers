```python
# a. Create a nave Bayes model that uses probability density functions to model the descriptive
# features in this dataset (assume that all the descriptive features are normally distributed).

from sklearn.naive_bayes import GaussianNB as naive_bayes_classifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd


#Reading the CSV File
df = pd.read_csv("C:/MAE551_HW6_Prob2/datafile.csv")
print(df)
print()

#Features matrix of the data
X = df[['SS-IN', 'SED-IN', 'COND-IN', 'SS-OUT', 'SED-OUT', 'COND-OUT']]
#Label
y = df['STATUS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10) 

#training the model according to features and labels
Model = naive_bayes_classifier() 
Model.fit(X_train, y_train) 

#Predictions of the model
y_pred = Model.predict(X_test) 

print("Model Accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
plt.plot(y_test,y_pred)
```

```python
# b. What prediction will the naive Bayes model return for the following query?

print(Model.predict([[222, 4.5, 1518, 74, 0.25, 1642]]))
```
